```python
from copy import deepcopy
stoneLocation=[2, 8]  #[row, column]
positiveLocation = [0,9,2]  #[row, column, utility value]
negativeLocation = [1,9, -2]  #[row, column, utility value]
iteration=20
noise=0.15
discount=0.91
grid_size =  10

grid = [0.0]*grid_size
for ix in range(grid_size):
    grid[ix]=[0.0]*grid_size
    
    

gridDirection = ['n']*grid_size
for ix in range(grid_size):
    gridDirection[ix]=['n']*grid_size

grid[stoneLocation[0]][stoneLocation[1]]= 'stone'
gridDirection[stoneLocation[0]][stoneLocation[1]] = 'None'

grid[positiveLocation[0]][positiveLocation[1]] = positiveLocation[2]
grid[negativeLocation[0]][negativeLocation[1]] = negativeLocation[2]
row_start = 0
row_end = len(grid)-1
col_start = 0
col_end = len(grid[0])-1

def calculation(dirn,row,col,grid_copy,discount,noise):
    if row == row_start and col == col_start:
        u_n = grid_copy[row][col]
        u_w = grid_copy[row][col]
            
        if grid_copy[row+1][col] == 'stone':
            u_s = grid_copy[row][col]
        else:
            u_s = grid_copy[row+1][col]
                
        if grid_copy[row][col+1] == 'stone':
            u_e = grid_copy[row][col]
        else:
            u_e = grid_copy[row][col+1]
    elif row == row_start and col == col_end:
        u_n = grid_copy[row][col]
            
        if grid_copy[row][col-1] == 'stone':
            u_w = grid_copy[row][col]
        else:
            u_w = grid_copy[row][col-1]
                
        if grid_copy[row+1][col] == 'stone':
            u_s = grid_copy[row][col]
        else:
            u_s = grid_copy[row+1][col]
                
        u_e = grid_copy[row][col]
            
    elif row == row_end and col == col_start:
        if grid_copy[row-1][col] == 'stone':
            u_n = grid_copy[row][col]
        else:
            u_n = grid_copy[row-1][col]
                
        u_w = grid_copy[row][col]
        
        u_s = grid_copy[row][col]
                
        if grid_copy[row][col+1] == 'stone':
            u_e = grid_copy[row][col]
        else:
            u_e = grid_copy[row][col+1]
                
    elif row == row_end and col == col_end:
        if grid_copy[row-1][col] == 'stone':
            u_n = grid_copy[row][col]
        else:
            u_n = grid_copy[row-1][col]
            
        if grid_copy[row][col-1] == 'stone':
            u_w = grid_copy[row][col]
        else:
            u_w = grid_copy[row][col-1]
            
        u_s = grid_copy[row][col]
                
        u_e = grid_copy[row][col]
            
    elif row == row_start:
        u_n = grid_copy[row][col]
            
        if grid_copy[row][col-1] == 'stone':
            u_w = grid_copy[row][col]
        else:
            u_w = grid_copy[row][col-1]
                
        if grid_copy[row+1][col] == 'stone':
            u_s = grid_copy[row][col]
        else:
            u_s = grid_copy[row+1][col]
                
        if grid_copy[row][col+1] == 'stone':
            u_e = grid_copy[row][col]
        else:
            u_e = grid_copy[row][col+1]
                
    elif col == col_start:
        if grid_copy[row-1][col] == 'stone':
            u_n = grid_copy[row][col]
        else:
            u_n = grid_copy[row-1][col]
            
        u_w = grid_copy[row][col]
                
        if grid_copy[row+1][col] == 'stone':
            u_s = grid_copy[row][col]
        else:
            u_s = grid_copy[row+1][col]
                
        if grid_copy[row][col+1] == 'stone':
            u_e = grid_copy[row][col]
        else:
            u_e = grid_copy[row][col+1]
                
    elif row == row_end:
        if grid_copy[row-1][col] == 'stone':
            u_n = grid_copy[row][col]
        else:
            u_n = grid_copy[row-1][col]
            
        if grid_copy[row][col-1] == 'stone':
            u_w = grid_copy[row][col]
        else:
            u_w = grid_copy[row][col-1]
                
        u_s = grid_copy[row][col]
                
        if grid_copy[row][col+1] == 'stone':
            u_e = grid_copy[row][col]
        else:
            u_e = grid_copy[row][col+1]
                
    elif col == col_end:
        if grid_copy[row-1][col] == 'stone':
            u_n = grid_copy[row][col]
        else:
            u_n = grid_copy[row-1][col]
            
        if grid_copy[row][col-1] == 'stone':
            u_w = grid_copy[row][col]
        else:
            u_w = grid_copy[row][col-1]
                
        if grid_copy[row+1][col] == 'stone':
            u_s = grid_copy[row][col]
        else:
            u_s = grid_copy[row+1][col]
                
        u_e = grid_copy[row][col]
            
    else:
        if grid_copy[row-1][col] == 'stone':
            u_n = grid_copy[row][col]
        else:
            u_n = grid_copy[row-1][col]
                
        if grid_copy[row][col-1] == 'stone':
            u_w = grid_copy[row][col]
        else:
            u_w = grid_copy[row][col-1]
                
        if grid_copy[row+1][col] == 'stone':
            u_s = grid_copy[row][col]
        else:
            u_s = grid_copy[row+1][col]
                
        if grid_copy[row][col+1] == 'stone':
            u_e = grid_copy[row][col]
        else:
            u_e = grid_copy[row][col+1]

                
    
    if dirn == 'n':    
        
        summ = u_n * (1-noise) + u_w *(noise/3) + u_e*(noise/3) + u_s *(noise/3)
        ex_utility = discount * summ
        
        return ex_utility,'n'
    
    elif dirn == 'w':
     
        summ = u_n * (noise/3) + u_w *(1-noise) + u_e*(noise/3) + u_s *(noise/3)
        ex_utility = discount * summ
        
        return ex_utility,'w'
    
    if dirn == 's':
            
        summ = u_n * (noise/3) + u_w *(noise/3) + u_e*(noise/3) + u_s *(1-noise)
        ex_utility = discount * summ
        
        return ex_utility,'s'
    
    elif dirn == 'e':
            
        summ = u_n * (noise/3) + u_w *(noise/3) + u_e*(1-noise) + u_s *(noise/3)
        ex_utility = discount * summ
        
        return ex_utility,'e'
    
def print_grid():
    for row in range(row_end+1):
        for col in range(col_end+1):
            if grid[row][col] == 'stone':
                print('(',grid[row][col],') ',end ='')
            else:
                print(grid[row][col],'(',gridDirection[row][col],')',end='')
        print()
    print()
print('Iteration # 0')
print_grid()

def start():
    global grid
    for i in range(iteration):
        print('Iteration #',i+1)
        grid_cop = deepcopy(grid)
        for row in range(row_end+1):
            for col in range(col_end+1):
                if (grid_cop[row][col] == positiveLocation[2]) or (grid_cop[row][col] == negativeLocation[2]) or (grid_cop[row][col] == 'stone'):
                    continue
                else:
                    temp_list = []
                    temp_val,dirn = calculation(dirn='n',row=row,col=col,grid_copy=grid,discount=discount,noise=noise)
                    temp_list.append((temp_val,dirn))
                    temp_val,dirn = calculation(dirn='w',row=row,col=col,grid_copy=grid,discount=discount,noise=noise)
                    temp_list.append((temp_val,dirn))
                    temp_val,dirn = calculation(dirn='e',row=row,col=col,grid_copy=grid,discount=discount,noise=noise)
                    temp_list.append((temp_val,dirn))
                    temp_val,dirn = calculation(dirn='s',row=row,col=col,grid_copy=grid,discount=discount,noise=noise)
                    temp_list.append((temp_val,dirn))

                    ex_utility =   max(temp_list,key = lambda item:item[0])
                    if ex_utility[0] > 0:
                        grid_cop[row][col] = round(ex_utility[0],2)
                        gridDirection[row][col] = ex_utility[1]

        grid = grid_cop
        print_grid()

start()

        
```
